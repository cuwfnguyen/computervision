```python
import cv2
import imutils
import numpy as np
import matplotlib.pyplot as plt
```

```python
I = cv2.imread('obj1.jpg', 0)
#gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(I, cmap='gray')
```

```python
output_adapthresh = cv2.adaptiveThreshold (I, 255.0, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2) 
plt.imshow(output_adapthresh,cmap="gray")
```

```python
img_blur=cv2.medianBlur(output_adapthresh,7)
plt.imshow(img_blur,cmap="gray")
```

```python
canny = cv2.Canny(img_blur,50, 150, 3)
plt.imshow(canny, cmap='gray')
```

```python
kernel = np.ones((12, 12), 'uint8')
dilate_img = cv2.dilate(canny, kernel, iterations=1)
plt.imshow(dilate_img, cmap='gray')
```

```python
canny1=dilate_img.copy()
cnts, h = cv2.findContours(canny1, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(canny1, cnts, -1, (255,0,0), 2)
plt.imshow(canny1)
print(len(cnts))
```
