```python
import cv2
import imutils
import numpy as np
import matplotlib.pyplot as plt
```

```python
I = cv2.imread('4.png', 0)
#gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(I, cmap='gray')
```

```python
output_adapthresh = cv2.adaptiveThreshold (I, 255.0,
cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, blockSize=91, C=-2) 
plt.imshow(output_adapthresh, cmap="gray")
```

```python
canny = cv2.Canny(output_adapthresh,50, 150, 3)
plt.imshow(canny, cmap='gray')
```

```python
cnts, h = cv2.findContours(canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
img_final= cv2.drawContours(I, cnts, -1, (255,255,0), 2)
plt.imshow(img_final)
print(len(cnts))
```
