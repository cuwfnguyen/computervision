```python
import cv2
import imutils
import numpy as np
import matplotlib.pyplot as plt
```

```python
I = cv2.imread('2.png', 0)
#gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(I,cmap="gray")
```

```python
img_blur= cv2.medianBlur(I,5)
plt.imshow(img_blur,cmap="gray")
```

```python
kernel = np.ones((9,5),np.uint8)
output_erosion = cv2.erode(img_blur, kernel)
plt.imshow(output_erosion, cmap="gray")
```

```python
blur=cv2.medianBlur(output_erosion,5)
plt.imshow(blur, cmap="gray")
```

```python
canny = cv2.Canny(blur,50, 150, 3)
plt.imshow(canny, cmap='gray')
```

```python
out=canny.copy()
cnts, h = cv2.findContours(out, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(out, cnts, -1, (255,255,0), 2)
plt.imshow(out)
print(len(cnts))
```
