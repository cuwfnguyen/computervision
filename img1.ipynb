```python
import cv2
import imutils
import numpy as np
import matplotlib.pyplot as plt
```

```python
I = cv2.imread('1.png', 0)
#gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(I, cmap='gray')
```

```python
output_adapthresh = cv2.adaptiveThreshold (I, 255.0, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 51, -20.0)
plt.imshow(output_adapthresh)
```

```python
kernel = np.ones((5,3),np.uint8)
output_erosion = cv2.erode(output_adapthresh, kernel)
plt.imshow(output_erosion)
```

```python
blur=cv2.medianBlur(output_erosion,5)
plt.imshow(blur, cmap="gray")
```

```python
canny = cv2.Canny(blur,50, 150, 3)
plt.imshow(canny, cmap='gray')
```

```python
out=canny.copy()
cnts, h = cv2.findContours(out, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(out, cnts, -1, (255,0,0), 2)
plt.imshow(out)
print(len(cnts))
```
